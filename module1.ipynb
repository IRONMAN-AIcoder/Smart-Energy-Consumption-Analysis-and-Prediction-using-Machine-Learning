#### The scope of this project is to collect, organize and analyze smart home energy consumption data at the device level in order to understand energy usage patterns and improve efficiency. ### Converting time from unix second format to normal date and time format

```python
df['time']=pd.to_numeric(df['time'], errors='coerce')
df=df.dropna(subset=['time'])
df['time']=pd.to_datetime(df['time'], unit='s')
df=df.set_index('time').sort_index()
print(df.index[:5])
```

### Organizing energy readings by device, room and timestamp

```python
device_cols=[
    'Dishwasher [kW]',
    'Furnace 1 [kW]',
    'Furnace 2 [kW]',
    'Home office [kW]',
    'Fridge [kW]',
    'Wine cellar [kW]',
    'Garage door [kW]',
    'Kitchen 12 [kW]',
    'Kitchen 14 [kW]',
    'Kitchen 38 [kW]',
    'Barn [kW]',
    'Well [kW]',
    'Microwave [kW]',
    'Living room [kW]',
    'Solar [kW]'
]
device_cols=[c for c in device_cols if c in df.columns]
```

```python
room_map={
    'Dishwasher [kW]': 'Kitchen',
    'Fridge [kW]': 'Kitchen',
    'Microwave [kW]': 'Kitchen',
    'Kitchen 12 [kW]': 'Kitchen',
    'Kitchen 14 [kW]': 'Kitchen',
    'Kitchen 38 [kW]': 'Kitchen',
    'Home office [kW]': 'Home Office',
    'Living room [kW]': 'Living Room',
    'Wine cellar [kW]': 'Basement',
    'Garage door [kW]': 'Garage',
    'Barn [kW]': 'Outdoor',
    'Well [kW]': 'Outdoor',
    'Furnace 1 [kW]': 'Heating',
    'Furnace 2 [kW]': 'Heating',
    'Solar [kW]': 'Generation'
}
```

### Creating device-room-energy table

```python
df_devices=df[device_cols].copy()
df_devices_hourly=df_devices.resample('h').sum()
```

```python
df_long=df_devices_hourly.reset_index().melt(id_vars='time', var_name='device', value_name='energy_kW')
```

```python
df_long['room'] = df_long['device'].map(room_map).fillna('Unknown')
print(df_long.head(10))
```
